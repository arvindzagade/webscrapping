```python
import requests
from bs4 import BeautifulSoup
```

```python
url ='https://www.bbc.com/news/business-51706225#:~:text=Although%20it%20said%20that%20the,such%20as%20India%20and%20China.'
```

```python
response = requests.get(url)
```

```python
response
```

```python
#Parsing 
soup = BeautifulSoup(response.text,'html.parser')
```

```python
soup
```

```python
soup.find_all('p')
```

```python
# text is present in list of paras so need to iterate.
```

```python
paras = soup.find_all('p')
print(type(paras))
```

```python
# now iterating over paras
for p  in paras:
    print(p.text) 
```
