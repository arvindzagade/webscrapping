#Webscrapping Images

```python
import requests
import urllib.request
from bs4 import BeautifulSoup
```

```python
url = 'https://yts.mx/browse-movies'
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
frames = soup.find_all('div',{'class':'browse-movie-wrap col-xs-10 col-sm-4 col-md-5 col-lg-4'})
frames
```

```python
def download(photo_url):
    name = photo_url.split('/')
    name = name [-2] + name[-1]
    urllib.request.urlretrieve(photo_url,name)
    

frames = soup.find_all('div',{'class':'browse-movie-wrap col-xs-10 col-sm-4 col-md-5 col-lg-4'})
for item in frames:
    figure = item.find('figure')
    photo_url = figure.img['src']
    print(photo_url)
    download(photo_url)
       
```
