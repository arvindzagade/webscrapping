```python
import requests
from bs4 import BeautifulSoup
```

```python
url ='https://www.icc-cricket.com/rankings/mens/player-rankings/odi'
```

```python
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
soup
```

```python
table = soup.find('table',{'class':"table"})
print(table)
```

```python
rank =2
for i in table.findAll('td',{'class':'table-body__cell name'}):
    name = i.text.replace(" ","")
    name = i.text.replace("\n","")
    print(str(rank)+'\t'+name)
    rank = rank + 1
  
   
```
