```python
import requests
from bs4 import BeautifulSoup
import urllib.request
```

```python
url = 'https://stackoverflow.com/questions/5041008/how-to-find-elements-by-class'
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text,'html.parser')
print(soup.prettify)
```

```python
answers = soup.findAll('div',{'id':'answers'})
answers
```

```python
for ans in answers:
    votes = ans.find('div',{'class':'js-vote-count'}).text
    solution = ans.find('div',{'class':'post-text'}).text
    if int(votes) >10:
        print(votes)
        print(solution)
```
